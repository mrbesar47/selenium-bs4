```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.firefox.options import Options
from bs4 import BeautifulSoup
import pprint
from time import sleep

#options = Options()
#options.headless = True
browser = webdriver.Chrome(executable_path="./chromedriver")#, options = options)
browser.get('https://hunt.thedapplist.com/')

base_url = 'https://hunt.thedapplist.com/'

#print("Title: %s" % browser.title)
sleep(5)

htmlSource = browser.page_source

soup = BeautifulSoup(htmlSource, 'lxml')

#apps = soup.find_all('a', class_='px-4 py-3 border-top d-block text-gray')

app_list_url = []

for a in soup.find_all('a', href=True):
    
    if (soup.find('span', class_='State d-inline-block mr-2 mb-2 bg-green') or (soup.find('span', class_='State d-inline-block mr-2 mb-2 bg-purple'))):
        #print("Found the URL:", a['href'])
        url = base_url + a['href']
        app_list_url.append(url)
        
countActive = 0
countClosed = 0
for span in soup.find_all('span', class_='State d-inline-block mr-2 mb-2 bg-green'):
    countActive+=1
    
for span in soup.find_all('span', class_='State d-inline-block mr-2 mb-2 bg-purple'):
    countClosed+=1

    
pprint.pprint(app_list_url)

browser.quit()
```

```python
print(len(app_list_url))
print(countActive)
print(countClosed)
```

```python
master_report = []

browser = webdriver.Chrome(executable_path="./chromedriver")#, options = options)

for url in app_list_url[1:countActive+1]:
    app_dict = {}
    app_dict['name'] = []
    app_dict['status'] = []
    app_dict['desc'] = []
    app_dict['author'] = []
    app_dict['votes'] = []
    app_dict['voters_address'] = [] 
    
    browser.get(url)
    sleep(2)
    htmlSource = browser.page_source
    appinfo = BeautifulSoup(htmlSource, 'lxml')
    dapp = appinfo.find('div', class_='p-4')
    print('-'*70)
         
    name = dapp.find('div', class_='text-white h3').text
    print(name)
    app_dict['name'] = name

    status = 'Active'
    app_dict['status'] = status

    desc = dapp.find('div', class_='h4 mb-3').text
    app_dict['desc'] = desc

    author = appinfo.find('span', class_='float-right').text
    app_dict['author'] = author

    zero = 0
    app_dict['votes'] = zero
    app_dict['voters_address'] = zero
    
    master_report.append(app_dict)
    
for url in app_list_url[countActive+1:]:
    app_dict = {}
    app_dict['name'] = []
    app_dict['status'] = []
    app_dict['desc'] = []
    app_dict['author'] = []
    app_dict['votes'] = []
    app_dict['voters_address'] = [] 
    
    browser.get(url)
    sleep(2)
    htmlSource = browser.page_source
    appinfo = BeautifulSoup(htmlSource, 'lxml')
    dapp = appinfo.find('div', class_='p-4')
    print('-'*70)        

    name = dapp.find('div', class_='text-white h3').text
    print(name)
    app_dict['name'] = name

    status = 'Closed'
    app_dict['status'] = status

    desc = dapp.find('div', class_='h4 mb-3').text
    app_dict['desc'] = desc

    author = appinfo.find('span', class_='float-right').text
    app_dict['author'] = author
    
    try:
         votes = (appinfo.find('span', class_="Counter ml-1").text)
    except:
        votes = 0
    
    app_dict['votes'] = votes

    if (votes == 0):
        app_dict['voters_address'] = 0
    else:
        voter_address = appinfo.find_all('span', class_='column')
        for voter in voter_address:
            app_dict['voters_address'].append(voter.text)

        
    master_report.append(app_dict)
    
    
browser.quit()
```

```python
import pandas as pd
df = pd.DataFrame(master_report)
df
```

```python
df = df.drop(['name'], axis = 1)

df.columns = [['Status', 'Description of DApp', 'Author', 'Votes', 'Voters address']]

df
```
